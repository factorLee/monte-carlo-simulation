```
# Monte Carlo integration
import random as rd
import numpy as np
import os

fun = lambda x:1/x
xA = 1. # x의 구간 시작
xB = 2.718 # y의 구간 끝
yA = 0. # f(x)의 최소값
yB = 1.0 # f(x)의 최대값보다 큰 임의의 값. 클수록 정확해지고 작을수록 빨라진다. f(x)의 최대값보다 같지 않고 더 큰 수를 넣어야 한다.
AREA = np.abs((xA-xB)*(yA-yB)) # 구간으로 둘러싸인 영역의 넓이.


rd.seed(os.times())

it = 10000

counted = 0

for i in range(it):
    if fun(rd.uniform(xA, xB)) > rd.uniform(yA, yB): counted+=1

print(str(it)+" iteration, " + str(AREA*float(counted)/(float(it)))+" is integration result.")
```

```
import numpy as np
import math
from sympy import Integral, Symbol
check_list = np.arange(1, 5, 0.001)
x = Symbol('x')
f = 1/x
check_ok = []
for i in check_list:
    integral_value = Integral(f, (x, (1, i))).doit()
    if round(integral_value) == 1.0:
        check_ok.append(i)
print(check_ok)

#integral_value = Integral(f, (x, (1, 4.5))).doit()
#round(integral_value)
#print(integral_value)

```

```
# Monte Carlo integration Wls
import random as rd
import os

fun = lambda x:1/x
xA = 1. # x의 구간 시작
xB = 0 # y의 구간 끝
yA = 0. # f(x)의 최소값
yB = 1.0 # f(x)의 최대값보다 큰 임의의 값. 클수록 정확해지고 작을수록 빨라진다. f(x)의 최대값보다 같지 않고 더 큰 수를 넣어야 한다.

rd.seed(os.times())

it = 10000

euler_list = []

for xB in check_ok:
    AREA = np.abs((xA-xB)*(yA-yB)) # 구간으로 둘러싸인 영역의 넓이.
    counted = 0 
    for i in range(it):
        if fun(rd.uniform(xA, xB)) > rd.uniform(yA, yB): 
            counted+=1
    monte_integal = AREA*(float(counted)/float(it))
    if round(monte_integal, 3) == 1.0:
            euler_list.append(round(xB, 4))

print(euler_list)
```
